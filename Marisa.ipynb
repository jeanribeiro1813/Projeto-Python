```python
def page (pag):
    return requests.get('https://www.marisa.com.br/store-finder?q=&page={}&latitude=-23.586164999999998&longitude=-46.675164699999996'.format(pag)).json()
```

```python
p = requests.get('https://www.marisa.com.br/store-finder?q=&page=0&latitude=-23.586164999999998&longitude=-46.675164699999996').json()
```

```python
total=p['total']
```

```python
i = ''
n = 0
list_res = []
while len(list_res) != total :
    result = page (n)
    n += 1
    for row in result['data']:
        list_res.append('{}|{}|{}|{}|{}|{}|{}|{}'
               .format(row['name'],row['town'],row['postalCode'],row['line1'],row['line2'],row['latitude'],row['longitude'],row['phone'] ))



with open ('C:/Users/Jean/Documents/Python Scripts/DadosMarisa.csv', 'w') as m:
    for item in list_res:
        print(item + '\n')
        m.write(item + '\n')
```
